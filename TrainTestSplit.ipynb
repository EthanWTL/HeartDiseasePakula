```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/Cleveland Imputed Data.csv')
print(df)
train, test = train_test_split(df, test_size=0.3)
print(train)
print(test)
train.to_csv("TrainTestData/cle_train.csv")
test.to_csv("TrainTestData/cle_test.csv")
```
